### #1 Explicacin del dataset y carga de datos

Vamos a trabajar con el dataset "Wine Quality", que es un dataset en el que se recogen diversas caractersticas fisicoqumicas de las variades tinta y blanca del "Vinho verde" portugus. Las variables del dataset se describen a continuacin


1. **Acidez fija**: La mayora de los cidos involucrados con el vino son fijos o no voltiles (no se evaporan fcilmente).
2. **Acidez voltil**: La cantidad de cido actico en el vino, que en niveles demasiado altos puede llevar a un sabor desagradable a vinagre.
3. **cido ctrico**: Encontrado en pequeas cantidades, el cido ctrico puede aadir 'frescura' y sabor a los vinos.
4. **Azcar residual**: Es la cantidad de azcar restante despus de que se detiene la fermentacin, es raro encontrar vinos con menos de 1 gramo/litro y vinos con ms de 45 gramos/litro se consideran dulces.
5. **Cloruros**: La cantidad de sal en el vino.
6. **Dixido de azufre libre**: La porcin libre de SO2 existe en equilibrio entre el SO2 molecular (como un gas disuelto) y el ion bisulfito; previene el crecimiento microbiano y la oxidacin del vino.
7. **Dixido de azufre total**: Es la cantidad de formas libres y ligadas de S02; en bajas concentraciones, el SO2 es en su mayora indetectable en el vino, pero en concentraciones libres de ms de 50 ppm, el SO2 se hace evidente en la nariz y el sabor del vino.
8. **Densidad**: La densidad del vino es cercana a la del agua dependiendo del porcentaje de alcohol y contenido de azcar.
9. **pH**: Describe qu tan cido o bsico es un vino en una escala de 0 (muy cido) a 14 (muy bsico); la mayora de los vinos estn entre 3-4 en la escala de pH.
10. **Sulfatos**: Un aditivo del vino que puede contribuir a los niveles de dixido de azufre (SO2), que acta como un antimicrobiano y antioxidante.
11. **Alcohol**: El porcentaje de contenido de alcohol del vino.
12. **Calidad**: Una variable de salida (basada en datos sensoriales, puntuacin entre 0 y 10).
13. **Class**: Si es blanco o tinto.

```python
train_set = train_set.drop(["free sulfur dioxide"], axis=1)
test_set = test_set.drop(["free sulfur dioxide"], axis=1)
```

```python
features = features.drop(["free sulfur dioxide"])
```

```python
train_set["class"] = train_set["class"].apply(lambda row: 1 if row == "red" else 0)   # "positive" o "affirmative" o como se indique
test_set["class"] = test_set["class"].apply(lambda row: 1 if row == "red" else 0)
```

```python
train_set[features].hist(bins=30, edgecolor='black', color='red')
plt.tight_layout()
```

```python
# Con un bucle
features_transform = ["residual sugar","chlorides","total sulfur dioxide","density", "alcohol" ]
for col in features_transform:
    train_set[col] = train_set[col].apply(np.log)			# el 1, 2n  para evitar el logaritmo de 0
```

```python
# y Visualizacin
train_set[features_transform].hist(bins=30, edgecolor='black', color='red')
plt.tight_layout();
```

```python
scaler = StandardScaler()
train_set[features] = scaler.fit_transform(train_set[features])
test_set[features] = scaler.transform(test_set[features])
```

```python
X_train = train_set.drop(target_qual, axis=1)
y_train = train_set[target_qual]
X_test = test_set.drop(target_qual, axis=1)
y_test = test_set[target_qual]
```

```python
mapeo_cut = {
    3: "v3",
    4: "v4",
    5: "v5",
    6: "v6",
    7: "v7",
    8: "v8",
    9: "v9"
}
y_train = y_train.map(mapeo_cut)
y_test = y_test.map(mapeo_cut)
```

```python
log_reg = LogisticRegression(max_iter = 25000, class_weight="balanced")
log_reg.fit(X_train, y_train)
train_pred=log_reg.predict(X_train)
scores = cross_val_score(log_reg, X_train, y_train, cv=5, scoring='accuracy')
print("Mean accuracy:", scores.mean())
```

```python
tree_clas = DecisionTreeClassifier(max_depth=5, random_state=42, class_weight="balanced")
tree_clas.fit(X_train, y_train)
train_pred = tree_clas.predict(X_train)
scores = cross_val_score(tree_clas, X_train, y_train, cv=5, scoring='accuracy')
print("Mean accuracy:", scores.mean())
```

```python
rnd_clf = RandomForestClassifier(n_estimators=500, max_depth=4, random_state=42, class_weight="balanced")
rnd_clf.fit(X_train, y_train)
train_pred = rnd_clf.predict(X_train)
scores = cross_val_score(rnd_clf, X_train, y_train, cv=5, scoring='accuracy')
print("Mean accuracy:", scores.mean())
```

```python
knn = KNeighborsClassifier(n_neighbors=5)  		#Un n_neighbors bajo hace que sobreajuste: menos varianza (menos validation error), a costa de ms training error
knn.fit(X_train,y_train)
train_pred = knn.predict(X_train)
scores = cross_val_score(knn, X_train, y_train, cv=5, scoring='accuracy')
print("Mean accuracy:", scores.mean())
```

```python
smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)

# Entrenar el modelo con los datos sobremuestreados
modelo_smote = knn = KNeighborsClassifier(n_neighbors=5) 
modelo_smote.fit(X_train_smote, y_train_smote)

# Realizar predicciones y evaluar el modelo
y_pred_smote = modelo_smote.predict(X_test)
print(classification_report(y_test, y_pred_smote))

y_train_smote.value_counts(True)
```
