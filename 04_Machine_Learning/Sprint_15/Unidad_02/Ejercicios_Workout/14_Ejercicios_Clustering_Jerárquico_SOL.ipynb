### Ejercicio 9

Nos hemos adelantado en el ejercicio anterior. Es hora de construir el clustering tal como hemos visto en el workout, utiliza el "AgglomerativeClustering" de sklearn para hacer el agrupamiento con el nmero de clusteres elegido y asigna las etiquetas a cada instancia. (Recuerda que hemos usado el linkage a "average", no a "ward" como se suele emplear)

```python
clustering_model = AgglomerativeClustering(n_clusters=4, metric='euclidean', linkage='average')
clustering_model.fit(train_set[features])
train_set['Cluster'] = clustering_model.labels_
```

### Ejercicio 11

Igual que podemos intuir que el cluster "pequeo" es el de las instituiciones peor puntuadas, muestra los valores medios de las features para cada clster y da tu interpretacin de que es cada uno. Emplea un grfico comparativo.


```python
features_mean_by_cluster = train_set[features + ["Cluster"]].groupby('Cluster').mean()
features_mean_by_cluster['Freq'] = train_set['Cluster'].value_counts().sort_index()
```

```python
# Perfil de cada cluster
fig, ax = plt.subplots(figsize=(12, 6))
features_mean_by_cluster.iloc[:, :-1].plot(kind='bar', ax=ax)
ax.set_ylabel('Mean Values')
ax.set_title('Cluster Analysis')
for i, v in enumerate(features_mean_by_cluster['Freq']):
    ax.text(i, max(features_mean_by_cluster.iloc[:, :-1].max().max(), 5) + 1, str(v), ha='center', va='bottom')
sns.despine(left=True, bottom=True)
plt.show()
```

*Podramos hacer la siguiente interpretacin:*
*Cluster 0, escuelas de valoracin baja en general y con un coste medio-alto*
*Cluster 1, instituciones top en casi todos los niveles, con un coste medio-alto*
*Cluster 2, instituciones de nivel muy bajo de coste tambin muy bajo que intentan suplir algo con las instalaciones e infraestructuras*
*Cluster 3, instintuciones de gama media pero con costes altos para sus puntuaciones*
