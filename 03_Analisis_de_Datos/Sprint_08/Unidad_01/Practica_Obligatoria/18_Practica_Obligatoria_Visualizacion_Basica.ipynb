### #1.1

Carga el dataset de casas de California que está en la ruta "./data/california_cities.csv". Muestra parte de su contenido y su descripción. Deshazte de las filas con nulos.

```python
df_california = pd.read_csv("./data/california_cities.csv")

df_california
```

```python
df_california.info()
```

```python
mediana_elevation_m = df_california['elevation_m'].median()
mediana_elevation_ft = df_california['elevation_ft'].median()
mediana_area_total_sq_mi = df_california['area_total_sq_mi'].median()
mediana_area_water_sq_mi = df_california['area_water_sq_mi'].median()
mediana_area_total_km2 = df_california['area_total_km2'].median()
mediana_area_land_km2 = df_california['area_land_km2'].median()
mediana_area_water_km2 = df_california['area_water_km2'].median()
mediana_area_water_percent = df_california['area_water_percent'].median()

df_california['elevation_m'] = df_california['elevation_m'].fillna(mediana_elevation_m)
df_california['elevation_ft'] = df_california['elevation_ft'].fillna(mediana_elevation_ft)
df_california['area_total_sq_mi'] = df_california['area_total_sq_mi'].fillna(mediana_area_total_sq_mi)
df_california['area_water_sq_mi'] = df_california['area_water_sq_mi'].fillna(mediana_area_water_sq_mi)
df_california['area_total_km2'] = df_california['area_total_km2'].fillna(mediana_area_total_km2)
df_california['area_land_km2'] = df_california['area_land_km2'].fillna(mediana_area_land_km2)
df_california['area_water_km2'] = df_california['area_water_km2'].fillna(mediana_area_water_km2)
df_california['area_water_percent'] = df_california['area_water_percent'].fillna(mediana_area_water_percent)
```

```python
df_california.info()
```

### #1.2

Recrea la siguiente figura, para ello construye en apartados separados cada gráfico y luego haz un apartado en el que se junten los cuatro. Las figuras son:
1. Las ciudades más pobladas, con los numeros de las poblaciones (tendrás que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrón, con trasparencia)
2. Histograma del área total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand itc", tamaño 18, color azul acero)
3. El scatter de área de agua de un ciudad con su poblacion.
4. El boxplot de la altura de las ciudades.

```python
# Ordenar el DataFrame por la columna 'poblacion' en orden descendente
df_poblacion_desc = df_california.sort_values(by='population_total', ascending=False)

# Seleccionar las primeras 5 filas
top_5 = df_poblacion_desc.head(5)

# Convertir las filas seleccionadas en un diccionario
top_5_city = list(top_5['city'])
top_5_pobl = list(top_5['population_total']/1000)

#Pintar diagrama de barras

kwargs = dict(alpha=0.5)

ax = plt.axes()
ax.bar(top_5_city, top_5_pobl, color = ["blue","grey","red","olive", "brown"],**kwargs)
ax.set_title('Ciudades más pobladas de California')
ax.set_ylabel('Habitantes (miles)');

#Falta plt.text
```

```python
areas = df_california['area_total_km2']

font_dict = {
    "size":18,
    "family":"Bradley Hand",
    "color":"black"}

ax = plt.axes()

#Pintar histograma
ax.hist(areas, bins=50, color='steelblue', label="Hist Area")
ax.legend(loc="upper right")
ax.set_xlabel("Superficie $(km^2)$", fontdict=font_dict)
ax.set_ylabel("Número de ciudades", fontdict=font_dict);
```

```python
ax = plt.axes()

x = df_california['area_water_km2']
plt.xlabel("Área de agua en la ciudad") 

y = df_california['population_total']/1000
plt.ylabel("Población (miles hab.)")

plt.scatter(x, y, c='green', alpha=0.7)
ax.set_title('Población por km cuadrado de agua')

props_flecha = {"arrowstyle":'->',
                "connectionstyle":f'arc3,rad=0.'}

ax.annotate('Outliers',xy=(87,3800000), xytext=(60,3000000),arrowprops=props_flecha)
ax.annotate('Outliers',xy=(120,1350000), xytext=(60,3000000),arrowprops=props_flecha);
```

```python
datos = df_california['elevation_m']

box_props = {'color':'red'}
cap_props = {'color':'red'}

ax = plt.axes()
ax.boxplot(datos, vert=False,patch_artist=True)

ax.set_title('Diagrama de caja de la altitud de las ciudades en California')
ax.set_xlabel('Altitud (m)')


#patch.set_facecolor(color)

;
```
