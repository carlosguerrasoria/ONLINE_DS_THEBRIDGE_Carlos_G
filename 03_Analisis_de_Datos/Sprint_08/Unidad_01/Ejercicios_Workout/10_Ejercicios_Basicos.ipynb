### Ejercicio 1

Construye la siguiente figura con dos subplots utilizando el estilo MATLAB (es decir plt.\<funcion\>, no usando objetos `Axes`)

```python
#Estilo MATLAB - plt.subplot
fig = plt.figure(figsize=(11.5, 15))

x =  np.linspace(0,10,100)

plt.subplot(2, 2, 1)
plt.plot(x, np.sin(x))
plt.title("Seno")


plt.subplot(2, 2, 2)
plt.plot(x, np.cos(x))
plt.title("Coseno")
```

```python
#Orientado a objetos - ax.subplot
fig, ax = plt.subplots(2,2)
ax[0,0].plot(x, np.sin(x))
ax[0,1].plot(x, np.cos(x))
```

### Ejercicio 2

Reproduce la imagen de la figura usando tres subplots en una columna, utilizando objetos `Axes`. Las dos primeras son las funciones anteriores y la tercera es la tangente (np.tan)


### Ejercicio 3: Ajuste de Colores y Estilos de Lnea  

Modifica la primera figura para que se muestre de la siguiente manera

```python
#Estilo MATLAB - plt.subplot

fig = plt.figure(figsize=(5, 8))

fig.subplots_adjust(hspace = 0.5, wspace = 0.5)

x =  np.linspace(0,10,100)

plt.subplot(2, 2, 1)
plt.plot(x, np.sin(x), ls = "--", color = ('#FD017D'))


plt.subplot(2, 2, 2)
plt.plot(x, np.cos(x), ls = "-.", color = ('c'))

```

### Ejercicio 4

Reproduce la figura siguiente. Recuerda que las funciones seno hiperblico y coseno hiperblico se obtienen as:
* np.sinh, seno hiperblico;
* np.cosh, coseno hiperblico;


```python
#Estilo MATLAB - plt.subplot
fig = plt.figure(figsize=(5.4,4))
fig.subplots_adjust(hspace = 1, wspace = 1)

x =  np.linspace(0,10,100)

plt.subplot(2, 1, 1)
plt.plot(x, np.sinh(x), color = "r")
plt.title("Seno Hiperbolico")
plt.xlabel('x')
plt.ylabel('sinh(x)')


plt.subplot(2, 1, 2)
plt.plot(x, np.cosh(x), color = "b")
plt.title("Coseno Hiperbolico")
plt.xlabel('x')
plt.ylabel('cosh(x)')
```

### Ejercicio 5: Ajustando el Espaciado de Subplots

Reproduce la siguiente figura con cuatro subplots, ajusta el espaciado entre los subplots para evitar la superosicin (tal y como ves en la figura), emplea `subplots_adjust` y los parmetros hspace y wspace.

```python
#Orientado a objetos - ax.subplot
fig, ax = plt.subplots(2,2, figsize=(4.7,3.5))
fig.subplots_adjust(hspace = 0.5, wspace = 0.5)

ax[0,0].plot(x, np.sin(x))
ax[0,1].plot(x, np.cos(x))
ax[1,0].plot(x, np.tan(x))
ax[1,1].plot(x, np.log(x))
```


### Ejercicio 6: 

Reproduce la figura siguiente empleando slo el estilo MATLAB (plt.\<funcion>\) y creando solo el contenido de los dos primeros grficos (fila 1). Luego usa el estilo con mtodos (objetos `axes`) para aadir el tercer grfico.

```python
#Estilo MATLAB - plt.subplot

fig = plt.figure(figsize=(7,6))
x =  np.linspace(0,10,100)

plt.subplot(2, 1, 1) # filas, columnas, indice
plt.plot(x,np.cos(x), label = "Coseno")

plt.subplot(2, 1, 1) # filas, columnas, indice
plt.plot(x,np.sqrt(x),label = "Raz cuadrada")

plt.title("Coseno y raiz cuadrada")
plt.legend();

ax = plt.subplots(1,1,figsize=(6.8,2))
ax[1].plot(x, np.sin(x))
plt.title("Seno")
```
