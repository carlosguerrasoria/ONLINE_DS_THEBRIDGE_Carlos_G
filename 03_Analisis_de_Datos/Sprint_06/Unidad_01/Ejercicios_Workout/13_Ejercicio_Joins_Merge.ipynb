### INNER JOIN

Para terminar, vamos a practicar con el INNER JOIN (también JOIN directamente) que en consultas a bases de datos es quizá más frecuente que el LEFT, pero en el caso de Data Science terminarás haciendo más LEFT que INNER (ya que LEFT conserva todos los valores e INNER sólo los que cruzan entre las dos tablas)

Ahora cambiamos de tablas, queremos tener de los clientes que compraron los datos del empleado que les atendió, sólo sí les atendió algún empleado. Es decir no me interesan los clientes que no fueron atendidos por algún empleado o que no se les asignó ningún empleado, y no me interesan los empleados que no atendieron a clientes. 

Como solo quiero tener clientes asociados a empleados y empleados a clientes, no manda ninguna si no su intersección -> INNER JOIN
