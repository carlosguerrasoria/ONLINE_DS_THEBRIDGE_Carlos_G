### Ejercicio 4



Modifica la funcin de clculo de distancias que hicimos en el workout para que funcione con vectores de cualquier dimensin

````python
import math

def distancia_2d(vec1, vec2, precision = 2): # Esperamos vectores de 2 dimensiones, es decir "listas" con longitud 2
    suma = 0
    for indice_componente in range(2): # 2 por ser la cardinalidad (nmero de componentes del vector)
        suma += (vec1[indice_componente] - vec2[indice_componente])**2
    distancia = math.sqrt(suma)
    distancia = round(distancia, precision)
    return distancia
````

````python
def dot_product(vec1, vec2, precision = 2):
    producto = 0
    for indice in range(len(vec1)): # Importante slo podemos hacer producto escalar de vectores con el mismo nmero de componentes o dimensiones
        producto += vec1[indice] * vec2[indice]
    producto = round(producto, precision)
    return producto
````

````python
def norm_vec(vec, precision = 1):
    escalar = dot_product(vec,vec) # Esta es la magia de las funciones y los mtodos construir unos sobre otros :-), reusar, craft, craft
    resultado = math.sqrt(escalar)
    resultado = round(resultado, precision)
    return resultado
````
